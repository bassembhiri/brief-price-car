```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import pandas as pd 
import seaborn as sns
import sklearn as skl
from sklearn.linear_model import SGDRegressor 
from sklearn.preprocessing import PolynomialFeatures 
import pymysql
import sqlalchemy as sql
from sqlalchemy import create_engine
%matplotlib inline
```

```python
car_data = pd.read_csv('carData.csv')
```

```python
x=np.array(car_data["Year"])
y=np.array(car_data["Selling_Price"])
x=x.reshape(-1, 1)

plt.scatter(x, y) 
```

```python
model1 = SGDRegressor(max_iter=100, eta0=0.0001)
model1.fit(x,y) 
```

```python
print('Coeff R2 =', model1.score(x, y))
plt.scatter(x, y)
plt.plot(x, model1.predict(x), c='red',lw=3)
```

```python
model2 = SGDRegressor(max_iter=1000, eta0=0.0001)
model2.fit(x,y) 
print('Coeff R2 =', model2.score(x, y))
plt.scatter(x, y)
plt.plot(x, model2.predict(x), c='red',lw=3)
```

```python
model5= SGDRegressor(max_iter=10000, eta0=0.0001)
model5.fit(x,y) 
print('Coeff R2 =', model5.score(x, y))
plt.scatter(x, y)
plt.plot(x, model5.predict(x), c='red',lw=3)
```

```python
x=np.array(car_data["Year"])
y=np.array(car_data["Selling_Price"])
x=x.reshape(-1, 1)

plt.scatter(x, y)

```

```python
y=y**2
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x = poly_features.fit_transform(x)


plt.scatter(x[:,0], y) 
```

```python
modelp = SGDRegressor(max_iter=10000, eta0=0.1)
modelp.fit(x,y)
print('Coeff R2 =', modelp.score(x, y))

plt.scatter(x[:,0], y, marker='o')
plt.scatter(x[:,0], modelp.predict(x), c='red', marker='+') 
```

```python
x=np.array(car_data["Year"])
y=np.array(car_data["Selling_Price"])
x2=x.reshape(-1, 1)
model6=SGDRegressor(loss="huber",penalty='l2',max_iter=10000, eta0=0.01,learning_rate='adaptive')
model6.fit(x2,y) 
print('Coeff R2 =', model6.score(x2, y))
```

```python
p = np.polyfit(x,y,1)
```

```python
f = np.poly1d(p)                # So we can call f(x)
fig = plt.figure()
ax  = fig.add_subplot(111)
plt.plot(x, y, 'bo', label="Data")
plt.plot(x,f(x), 'r',label="Polyfit")
plt.plot(x2, model6.predict(x2), c='k',lw=3)
plt.show()
```

```python

slope, intercept, r_value, p_value, std_err = sc.stats.linregress(x, y)
```

```python
def predict(x):
   return slope * x + intercept
```

```python
z = predict(x)
```

```python
axes = plt.axes()
axes.grid() # dessiner une grille pour une meilleur lisibilit du graphe
plt.scatter(x,y) # X et Y sont les variables qu'on a extraite dans le paragraphe prcdent
plt.plot(x, z, c='r')
plt.plot(x,f(x), 'g',label="Polyfit")
plt.show()
```
